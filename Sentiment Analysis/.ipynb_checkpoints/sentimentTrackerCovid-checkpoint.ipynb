# News sentiment tracker.
This script is getting articles from the Guardian website and randomly selecting them to then see if they are positive or negative.  First what I did was download reviews in Europe from AirBnB.  I went with native English speakers as I wanted to train this model with native speakers.  I took European reviews because the Guardian is in Europe.  I randomly selected 50,000 reviews.  I was concerned that this could be trained to just recognize hotel lingo if you will, so I then decided to include 25,000 IMDB movie reviews.  Before combining them, I cleaned the data from any html formatting and then combined both and finally randomly selecting 25,000 airBnB and movie reviews.  Hopefully, this will give me a better outcome.  For some reason, the NLTK from class would not work on my computer as it was taking forever to train the model.  I found a better model online which I think as it gives me a higher accuracy than the one in class.  I used a count vectorizer instead of NLTK's tokenizer because it ignores stop words and punctuations. Secondly, I used TfidTransformer to adapt the text data by creating and returning a count vectorized output of the text.  This was quite accurate and runs very quickly this took minutes to run.  It gets rid of unneeded words and vocabulary so the machine can focus on the important text.  Term frequency is important too in the positive or negative sentiment outcome.  Took me a while to get this done but it is done!

```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import sklearn
import pickle
from textblob import TextBlob
import mysql.connector
```

## Get the review datase to train the model

```python
df = pd.read_excel("reviewData1962020.xlsx")
```

## View the dataset

```python
df.head()
```

## Verify dataset details

```python
df.info()
```

## Split the data to go into the model.
80% / 20% seems to be a good split of the data as the accuarcy with this is high.

```python
X_train, X_test, y_train, y_test = train_test_split(df.review, df.sentiment, test_size=0.2,random_state=12)
```

# Initialize Count Vectorizer
Minimum frequency is 2 and I will use the top 5000 words that are used in the dataset.

```python
reviewVectorizer= CountVectorizer(min_df=2, tokenizer=nltk.word_tokenize, max_features=5000)
```

## Fit and tranform reviews by using training text.

```python
train_counts = reviewVectorizer.fit_transform(X_train)
```

## View the training counts to verify job done.

```python
train_counts.shape
```

## Now transform the text and output a count vectorized text

```python
reviewTfmer = TfidfTransformer()
train_tfidf = reviewTfmer.fit_transform(train_counts)
```

## Again verify the training counts match above.

```python
train_tfidf.shape
```

## Using the fitted vectorizer and transformer
This will tranform the dataset text

```python
test_counts = reviewVectorizer.transform(X_test)
test_tfidf = reviewTfmer.transform(test_counts)
```

## Now for the fun part which is to Train the model
Using a Naive Bayes classifier to train the model.

```python
model = MultinomialNB()
model.fit(train_tfidf, y_train)
```

## Find the accuracy of the model
This is important to verify the accuracy of the model as it needs to be high to predict accurately all future text.

```python
y_pred = model.predict(test_tfidf)
print("Model accuracy score: ", sklearn.metrics.accuracy_score(y_test, y_pred))
```

## Confusion Matrix to plot the accuracy
This will give me an error matrix to visualize better the accuracy of the model.  I used a heatmap to visualize the confusion matrix.

```python
cm = confusion_matrix(y_test, y_pred)
classes = ['Positive', 'Negative']
plt.figure(figsize = (10, 10))
plt.imshow(cm, interpolation='nearest')
plt.title('Confusion Matrix of the Sentiment Analysis', size = 24)
plt.colorbar(aspect=4)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45, size = 14)
plt.yticks(tick_marks, classes, size = 14)
thresh = cm.max() / 2.

for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, format(cm[i, j]), fontsize = 20, horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
plt.grid(None)
plt.tight_layout()
plt.ylabel('Actual or True', size = 18)
plt.xlabel('Predicted', size = 18)
```

## Plot the percentage of positive or negative articles

```python
labels = 'Positive', 'Negative'
positive = newsArticles[newsArticles['Sentiment'].str.contains("positive")]
negative = newsArticles[newsArticles['Sentiment'].str.contains("negative")]

pos = positive.size
neg = negative.size
sizes = [pos,neg]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## A bit more analysis
Looking at the pie chart above it looks to be almost 50% / 50%.  I decided to add more articles to see if this would change and it did not.  I am not sure this would be true so I and to look sentiment polarity particularly polarity and subjectivity.  Let us get that data and then plot.  Did not think for sure that it could be more positive than negative.

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity
polarityList = [pol(x) for x in newsArticles.text]
subjectivityList = [sub(x) for x in newsArticles.text]
analysis = pd.DataFrame(polarityList,columns={'Polarity'})
analysis['Subjectivity'] = subjectivityList
```

## Now we have it let us plot it.
Seems almost 50/50 as well with a few outliers

```python
analysis.plot(style=['o','rx'])
```

## One more to be sure.
Very concentrated at the center showing it is almost 50% / 50%

```python
plt.figure(figsize=(8,6))
ax = sns.regplot(x="Polarity", y="Subjectivity",fit_reg=False, scatter_kws={'alpha':0.5},data=analysis)
plt.ylabel('Subjectivity Frequency')
plt.xlabel('Polarity Frequency')
plt.title('Polarity Frequency vs Subjectivity Frequency',fontsize=20)
```
