Now to create a class with a method putting in the individual constraints for this problem.  Importing from the CSP framework the nodes are brought in once a domain assignment is given and then to verify that the node coloring assignments are satisfied with the constraints.  Firstly, the Constraint method below checks if the neighboring nodes are in the assignment array.  Secondly it goes through each neighboring node in the assignment and verifies that the neighboring nodes do not have the same domain or color.  If the above conditions are met, then it returns false and, in the framework, inside the loop it goes to the next color or domain in domains till we do not have a conflict or run out of domains.  This is a very simple code but just took many iterations for me to get here.  As I said before the theory makes sense and now the code makes sense for me.

```python
class mapColoring(CSP):
    def constraints(self, neighbor, assignment, domain):
        for neighbor in self.variables[neighbor]:
            if neighbor in assignment and domain == assignment[neighbor]:
                return False
        return True
```

The idea is the same as the map above is that neighboring nodes do not have the same variable in this case a number.  Adding the variables needed below then going into a class that makes a graph.

```python
rows = ['1','2','3','4','5','6','7','8']
columns = ['A','B','C','D','E','F','G','H']
domains = ['1','2','3','4','5','6','7','8']
isGraph = True
```

The class graph is to bring in variables needed to create a graph with a method.  In the init method takes the row and column values and makes a list of all the possible variables in the 8X8 grid.  Using letters for columns and numbers for rows, using strings made it a bit easier as well. Once the grid is created then on to making the graph.
Firstly, I create a blank graph dictionary to add all the nodes(keys) with the adjacent nodes.  Two loops make this possible as we go through each variable, we then can add all the adjacent nodes to each node.  I used neighbors dictionary to capture the adjacent nodes or neighbors and nodes to create a key.  Looping and working through each variable and adding to the graph dictionary.  Making the columns letters instead of numbers needs to be this way to make the loops work giving neighbors with the same number(row) or same letter(column).  Once this is all done this will return a graph with nodes (keys) with neighboring nodes. Now onto constraints.

```python
class Graph():
    def __init__(self, rows, columns):
        self.rows = rows
        self.columns = columns
        self.variables = []
        for row in rows:
            for column in columns:
                self.variables.append(column+ ',' + row)

    def makeGraph(self):
        graph = {}
        for node in self.variables: 
            neighbors = {}
            nodes =  {}
            for node2 in self.variables:
                if node[0] in node2 or node[2] in node2:
                    if node != node2:
                        nodes[node2] = node2
                        neighbors[node] = nodes
            graph.update(neighbors)
        return graph
```

Below is the class that has the constraints verifying that each neighboring node does not have the same number.  In this case neighboring nodes are in the same row and column.

```python
class latinBox(CSP):
    def constraints(self, neighbor, assignment, domain):
        for neighbor in self.variables[neighbor]:
            if neighbor in assignment and domain == assignment[neighbor]:
                return False
        return True
```

Now to call each class.  First calling the graph class and then the method creating the graph with rows and columns. Finally, once the graph has been constructed then run it through the CSP recursive back tracking method and constraints method to make sure all constraints are met.  If they are met, then a solution is found if not no solution will be given.  In this algorithm I think the min conflicts method would work more efficient, but this was not part of the CA.  

```python
graph1 = Graph(rows, columns)
graph = graph1.makeGraph()
start = latinBox(graph, domains, isGraph)
solution = start.recursiveBackTracking(assignment = {})
```

Now to print out if a solution is found or not.  I had to create some loops to display the data that can be visually verified.  This was a bit fun figuring out the best way to do this.

```python
if solution is None:
    print("No solution found!")
else:
    variables = list(solution.keys())
    node = 0
    for i in range(len(columns)):
        lists = []
        for j in range(len(rows)):
            current = variables[node]
            lists.append(solution[current])
            node += 1
        print(lists)
```
